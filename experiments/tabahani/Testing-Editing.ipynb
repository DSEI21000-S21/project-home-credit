```python
df_application_train = pd.read_csv('application_train.csv')
df_application_train['SK_ID_CURR'] = df_application_train['SK_ID_CURR'].astype(str)
df_application_train = df_application_train[['SK_ID_CURR','TARGET']]
```

```python
df_credit_card_balance['SK_ID_CURR'] = df_credit_card_balance['SK_ID_CURR'].astype(str)
```

```python
# df_credit_card_balance.merge(df_application_train[['TARGET']], on='SK_ID_CURR',how='left')
```

```python
full = pd.merge(
    df_credit_card_balance,
    df_application_train, #here the differance
    on='SK_ID_CURR',
    how='left' #and add the how='left'
)
```

```python
full
```

```python
full['TARGET'].unique()
```

```python
full_not_nan = full.dropna()
```

```python
full_not_nan
```

```python
df_majority = full_not_nan[full_not_nan.TARGET==0]
df_minority = full_not_nan[full_not_nan.TARGET==1]

print(df_majority)
print(df_minority)
```

```python
print(len(df_majority))
print(len(df_minority))
```

```python
sizes = [len(df_majority),len(df_minority)]
labels = ['Case 0','Case 1']
fig1, ax1 = plt.subplots()
ax1.pie(sizes,labels=labels,autopct='%1.1f%%')
plt.show()
```

```python
from sklearn.utils import resample
df_majority_downsampled = resample(df_majority, 
                                 replace=False,    # sample without replacement
                                 n_samples=len(df_minority),     # to match minority class
                                 random_state=123) 
 
# Combine minority class with downsampled majority class
df_downsampled = pd.concat([df_majority_downsampled, df_minority])
df_downsampled
```

```python

x_values = df_downsampled[plotting_card_cols]
y_values = df_downsampled['TARGET']
```

```python
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.linear_model import LogisticRegression

X_train, X_test, y_train, y_test = train_test_split(x_values, y_values, test_size=0.33, random_state=42)
```

```python
logit = LogisticRegression(random_state=0,solver='sag').fit(X_train,y_train)
svmModel = svm.LinearSVC().fit(X_train,y_train)
```

```python
log_pred = logit.predict(X_test)
svm_pred = svmModel.predict(X_test)

```

```python
from sklearn.metrics import accuracy_score,classification_report
print("Logistic")
print(classification_report(y_test,log_pred))
print(accuracy_score(y_test,log_pred))
```

```python
print("SVM")
print(classification_report(y_test,svm_pred))
print(accuracy_score(y_test,svm_pred))
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_values_transform = scaler.fit_transform(x_values)
```

```python
X_train_scale, X_test_scale, y_train_scale, y_test_scale = train_test_split(x_values_transform, y_values, test_size=0.33, random_state=42)
```

```python
logit_scale = LogisticRegression(random_state=0,solver='sag').fit(X_train_scale,y_train)
svmModel_scale = svm.LinearSVC().fit(X_train_scale,y_train)
```

```python
log_pred_scale = logit_scale.predict(X_test)
svm_pred_scale = svmModel_scale.predict(X_test)
```

```python
print("Logistic")
print(classification_report(y_test,log_pred_scale))
print(accuracy_score(y_test,log_pred_scale))
```

```python
print("SVM")
print(classification_report(y_test,svm_pred_scale))
print(accuracy_score(y_test,svm_pred_scale))
```
